```python
import tkinter as tk
from tkinter import ttk

# Create the main window
root = tk.Tk()
root.title("GUI with Dropdown, Window, and Subwindows")

# Get the screen dimensions
screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()

# Set the size and position of the main window
window_width = int(screen_width * 0.8)
window_height = int(screen_height * 0.8)
window_x = int((screen_width - window_width) / 2)
window_y = int((screen_height - window_height) / 2)
root.geometry(f"{window_width}x{window_height}+{window_x}+{window_y}")

# Create the dropdown menu in the top left corner
menu_frame = ttk.Frame(root, padding=10)
menu_frame.pack(side="top", fill="x", padx=10, pady=10)
menu_label = ttk.Label(menu_frame, text="Select an option:")
menu_label.pack(side="left")
menu_var = tk.StringVar()
menu_var.set("Option 1")
menu_dropdown = ttk.OptionMenu(menu_frame, menu_var, "Option 1", "Option 2", "Option 3")
menu_dropdown.pack(side="left", padx=10)

# Create the window in the top center section
window_frame = ttk.Frame(root, padding=10)
window_frame.pack(side="top", fill="both", expand=True, padx=10, pady=10)
window_label = ttk.Label(window_frame, text="Window")
window_label.pack(side="top", anchor="center")
window_textbox = tk.Text(window_frame)
window_textbox.pack(side="top", fill="both", expand=True)

# Create the subwindows in the bottom section
subwindows_frame = ttk.Frame(root, padding=10)
subwindows_frame.pack(side="bottom", fill="both", expand=True, padx=10, pady=10)

subwindow1 = ttk.Frame(subwindows_frame, relief="solid", borderwidth=2, padding=10)
subwindow1.grid(row=0, column=0, padx=10, pady=10, sticky="nsew")
subwindow1_label = ttk.Label(subwindow1, text="Subwindow 1")
subwindow1_label.pack(side="top", padx=10, pady=10)
subwindow1_textbox = tk.Text(subwindow1)
subwindow1_textbox.pack(side="top", fill="both", expand=True, padx=10, pady=10)

subwindow2 = ttk.Frame(subwindows_frame, relief="solid", borderwidth=2, padding=10)
subwindow2.grid(row=0, column=1, padx=10, pady=10, sticky="nsew")
subwindow2_label = ttk.Label(subwindow2, text="Subwindow 2")
subwindow2_label.pack(side="top", padx=10, pady=10)
subwindow2_textbox = tk.Text(subwindow2)
subwindow2_textbox.pack(side="top", fill="both", expand=True, padx=10, pady=10)

subwindow3 = ttk.Frame(subwindows_frame, relief="solid", borderwidth=2, padding=10)
subwindow3.grid(row=0, column=2, padx=10, pady=10, sticky="nsew")
subwindow3_label = ttk.Label(subwindow3, text="Subwindow 3")
subwindow3_label.pack(side="top", padx=10, pady=10)
subwindow3_textbox = tk.Text(subwindow3)
subwindow3_textbox.pack(side="top", fill="both", expand=True, padx=10, pady=10)

subwindows_frame.rowconfigure(0, weight=1)
subwindows_frame.columnconfigure(0, weight=1)
subwindows_frame.columnconfigure(1, weight=1)
subwindows_frame.columnconfigure(2, weight=1)

root.resizable(True, True)

menu_frame.pack_propagate(False)
window_frame.pack_propagate(False)
subwindow1.pack_propagate(False)
subwindow2.pack_propagate(False)
subwindow3.pack_propagate(False)
subwindows_frame.pack_propagate(False)

# Run the main loop
root.mainloop()
```
